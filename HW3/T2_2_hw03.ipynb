## Problem 1

Gaussian elimination can also be used to find solutions to linear equations of the form $Ax=b$ where $A\in\mathbb{R}^{m\times n}$ and $b\in\mathbb{R}^n$. One simply forms the **augmented matrix**


$$
\left(\begin{array}{c|c} A & b\end{array}\right) = \left(\begin{array}{cccc|c}a_{1,1} & a_{1,2}&\cdots &a_{1,n} & b_1\\
a_{2,1} & a_{2,2}&\cdots &a_{2,n} & b_2\\
\vdots & \vdots & \ddots & \vdots & \vdots\\
a_{m,1} & a_{m,2} &\cdots & a_{m,n} & b_m
\end{array}\right).
$$

and performs Gaussian elimination until the left part of the augmented matrix is in row echelon form. For convenience, we will write this row echelon form of $A$ as $REF(A)$.

Moreover, Gauss-Jordan elimination allows us to write down a **parameterization** of the solution set to $Ax=b$ in the form

$$
x^\prime + \sum_{i\in\mathcal{F}} x_i v_i
$$

$v_i\in\mathbb{R}^n$, $x_i$ is a free variable, and $i\in\mathcal{F}$ if and only if $REF(A)$ does not have a pivot in column $i$. That is, $\mathcal{F}$ stores all the indices of the free variables. 

Gauss-Jordan elimination begins by performing Gaussian elimination to get a matrix in row echelon form. In the next step, each pivot row is multiplied by the reciprocal of the leading entry until all pivot entries equal $1$. In the final step, starting with the left-most pivot, all the entries above each pivot are eliminated by adding the appropriate multiple of the pivot row. The resulting matrix is said to be in **reduced row echelon form**, and we will sometimes write $RREF(A)$ to denote this matrix.

After all this is done (remembering that the row operations will be applied to the augmented column as well), we will have an augmented matrix of the form

$$
\left(\begin{array}{c|c} RREF(A) & \tilde{b}\end{array}\right).
$$

We can read off the parameterization of the solution set as follows: if $x\in\mathbb{R}^n$, then $x_i=x_i$ if $i$ is the index of a free variable (i.e. the index of a column with no pivot), and if $i$ has a pivot in row $k$, we set

$$
x_i = \tilde{b}_k - \sum_{j\in\mathcal{F}} x_j v_{i,\:j}
$$

where $v_{i,j}$ is the entry of $RREF(A)$ in the $i$th row and $j$th column. Letting $v_j$ be the vector with entries

$$
v_{i,\:j} = \left\{\begin{array}{cl} -1 &\text{ if }i=j\\ 0 &\text{ if }i\in\mathcal{F},\: i\not=j\\ v_{i,\: j} & \text{ if } i\not\in\mathcal{F}\end{array}\right.,
$$

and let $x^\prime$ be the vector with entries $x_i^\prime=0$ if $i\in\mathcal{F}$ and $x_i^\prime=\tilde{b}_k$ if $i\not\in\mathcal{F}$, where $k$ is the row index of the pivot in column $i$. Then

$$
x^\prime + \sum_{i\in\mathcal{F}} x_i v_i.
$$

## Problem 4

Since it can be difficult or impossible to find an exact expression for the largest eigenvalue of a matrix, we must rely on numerical methods. One way to compute an approximation to the largest eigenvalue and its eigenvector is via a **power iteration**: if $A\in\mathbb{R}^{n\times n}$ and $Av_0\not=0$ for $v^{(0)}\in\mathbb{R}^n$, we recursively set

$$
v^{(k+1)} = \frac{1}{\Vert Av^{(k)}\Vert}Av^{(k)}.
$$

The idea is that, if $v= x+h$ with $x$ an eigenvector of $A$ corresponding to the largest eigenvalue, and $h$ is orthogonal to $x$, then $Av=Ax+Ah=\lambda x+Ah$, and we are growing the magnitude of the term along the eigenvector.

### Part A

```python
import numpy as np
import pandas as pd

from sklearn.datasets import load_boston
from sklearn.preprocessing import scale
from sklearn.linear_model import LinearRegression

import matplotlib.pyplot as plt

boston=load_boston()
#print(boston.feature_names)
df = pd.DataFrame(boston.data, columns=boston.feature_names)
df['target'] = pd.Series(boston.target)
#print(df.loc[0:6,])
df=df.loc[:,['NOX','target']]
X=scale(df)
#print(X[:,1])
reg=LinearRegression().fit(np.array([X[:,0]]).reshape(506,1),X[:,1])
fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)
ax1.scatter(X[:,0], X[:,1],color='purple')
ax1.plot(X[:,0],reg.coef_*X[:,0]+reg.intercept_)
plt.show()
#plt.plot([X[:,0]],X[:1])
```

### Part B
We can observe that the logarithmic value of the norm of bias vector at each step has a linear relation with the step.

```python
A=(np.mat(X).T)*np.mat(X)
norm_V=np.arange(1,21,dtype='f')
v0=np.array([1,0],dtype='f').reshape(2,1)
v=np.array([0,0],dtype='f').reshape(2,1)
for i in range(1,21):
    v=A*v0
    v=v/np.linalg.norm(v)
    #print(np.linalg.norm(v-v0))
    norm_V[i-1]=np.linalg.norm(v-v0)
    v0=v
#print(norm_V)
plt.semilogy(np.array(range(1,21)),norm_V)
```
