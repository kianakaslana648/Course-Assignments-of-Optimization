## Problem 1

In lecture, we demonstrated that a matrix of the form $XX^T$ has a factorization as $U\Lambda U^T$ where the columns of $U$ form an orthonormal collection and $\Lambda$ is a diagonal matrix with strictly positive entries on the diagonal. In this problem, we will prove that a generic symmetric $n$ by $n$ matrix is diagonalizable. 

Suppose $A$ is an $n$ by $n$ symmetric matrix, and let $A=U\Sigma V^T$ be the singular value decomposition of $A$. In particular, we note that $A=UU^TA$ by our construction. Assume that all the singular values of $A$ are unique and strictly positive.

### Part A

Show that $AV=U\Sigma$ and $AU=V\Sigma^T$, and use this to show that $col(A)=col(U)=col(V)$.

### Part B

Use Part A to show that $UU^TV=V$.

### Part C

Show that $\Sigma U^TV= U^TV \Sigma$, and use this to conclude that the columns of $U^TV$ are all eigenvectors of $\Sigma$. Since $\Sigma$ is diagonal with unique eigenvalues, we must conclude that $U^TV$ is a diagonal matrix with $\pm 1$ entries on the diagonal.

### Part D

Use Part B and Part C to show that $A = V \Lambda V^T$ where $\Lambda= V^TU\Sigma$. 


### Note: A general diagonalization of a symmetric matrix has the form $A=U\Lambda U^T$ where $U$ is an orthogonal matrix and some diagonal entries of $\Lambda$ may be zero.

## Solution to Problem 1
### Part A
$$AV=A(\Sigma^{-1}U^{-1}A)^T=AA^T(U^{-1})^{T}(\Sigma^{-1})^{T} $$
$$=(UU^TA)A^TU\Sigma^{-1}=U(U^TAA^TU)\Sigma^{-1}$$
$$=U\Sigma^2\Sigma^{-1}=U\Sigma$$
$$V\Sigma^T=(\Sigma^{-1}U^{-1}A)^T\Sigma^T=A^T(U^{-1})^{T}(\Sigma^{-1})^{T}\Sigma^T$$
$$=AU$$
From the first equation, we have $(row(A),col(V))=(row(U),col(\Sigma))$  
From the second equation, we have $(row(A),col(U))=(row(V),row(\Sigma))$
$$col(V)=col(\Sigma)=row(\Sigma)=col(U)=row(U)=row(A)=col(A)$$
### Part B
In fact $row(U)=col(U)=col(V)$ and we have:
$$U^TU=I\rightarrow U^T=U^{-1}\rightarrow UU^T=I$$
So $UU^TV=V$
### Part C
#### 1st part
If we write $U$ as $(u_1,u_2,...,u_n)$ and $V$ as $(v_1,v_2,...,v_n)$, in which $\{u_i\}_{i=1}^{n}$ and $\{v_i\}_{i=1}^{n}$ represent the column vectors of $U$ and $V$. Since $A$ is symmetric, we have:
$$U\Sigma V^T=V\Sigma U^T$$
$$(\sigma_1u_1,\sigma_2u_2,...,\sigma_nu_n)\begin{pmatrix}
v_1^T\\
v_2^T\\
\vdots\\
v_n^T
\end{pmatrix}=(\sigma_1v_1,\sigma_2v_2,...,\sigma_nv_n)\begin{pmatrix}
u_1^T\\
u_2^T\\
\vdots\\
u_n^T
\end{pmatrix}$$
$$\sum_{i=1}^{n}\sigma_iu_iv_i^T=\sum_{i=1}^{n}\sigma_iv_iu_i^T$$
$$u_k^T(\sum_{i=1}^{n}\sigma_iu_iv_i^T)u_k=u_k^T(\sum_{i=1}^{n}\sigma_iv_iu_i^T)u_k,\ k=1,2,...,n$$
$$\sigma_k(u_k^Tu_k)v_k^Tu_k=\sigma_ku_k^Tv_k(u_k^Tu_k),\ k=1,2,...,n$$
$$v_k^Tu_k=u_k^Tv_k,\ k=1,2,...,n$$
That is:
$$V^TU=U^TV$$
Since $A$ is symmetric, we have:
$$U\Sigma V^T=V\Sigma U^T$$
$$\Sigma V^TU=U^TV\Sigma,\ (Multiply\ by\ U^T\ on\ the\ left,\ U\ on\ the\ right)$$
We have:
$$U^TV\Sigma=\Sigma V^TU=\Sigma U^TV$$
#### 2nd part
We can write $U^TV$ as $(w_1,w_2,...,w_n)$, in which $\{w_i\}_{i=1}^{n}$ are the column vectors of $U^TV$. And we write $\Sigma$ as:
$$\begin{pmatrix}
\sigma_1&0&{\cdots}&{0}\\
0&\sigma_2&{\cdots}&{0}\\
{\vdots}&{\vdots}&{\ddots}&{\vdots}\\
{0}&{0}&{\cdots}&{\sigma_n}
\end{pmatrix}$$
Then we have:  
$$\Sigma (w_1,w_2,...,w_n)=(w_1,w_2,...,w_n)\begin{pmatrix}
\sigma_1&0&{\cdots}&{0}\\
0&\sigma_2&{\cdots}&{0}\\
{\vdots}&{\vdots}&{\ddots}&{\vdots}\\
{0}&{0}&{\cdots}&{\sigma_n}
\end{pmatrix}$$
$$(\Sigma w_1,\Sigma w_2,...,\Sigma w_n)=(\sigma_1w_1,\sigma_2w_2,...,\sigma_nw_n)$$
That is:
$$\Sigma w_i=\sigma_iw_i,\ for\ i\ \in\{1,2,...,n\}$$
#### 3rd part
The element of $\Sigma U^TV$ at the position of row $i$ and column $j$ is $\sigma_i u_i^Tv_j$; the element of $U^TV\Sigma$ at the position of row $i$ and column $j$ is $\sigma_j u_i^Tv_j$. We have:
$$\sigma_i u_i^Tv_j=\sigma_j u_i^Tv_j$$
for $(i,j)\ that\ i\neq j$:
$$u_i^Tv_j=0$$
For each $u_i,v_i$, it's orthogonal to other $n-1$ orthonormal vectors $\{v_1,...v_{i-1},v_{i+1},...v_n\}$. Since $R^n$ is only n-dimensional, the only left space for $u_i,v_i$ is a line. They both have length of $1$. So $u_i=v_i\ or\ u_i=-v_i$. That leads to: $u_i^Tv_i=1\ or\ u_i^Tv_i=-1$. 
In summary, the diagonal elements of $U^TV$ belong to $\{-1,1\}$, the elements at other positions are all zero.
### Part D
$A = V \Lambda V^T\Leftrightarrow A = VV^TU\Sigma V^T\Leftrightarrow A=U\Sigma V^T$

## Problem 2

Diagonalization allows us to perform convegence analysis of the power iteration for numerical construction of an eigenvector for a symmetric matrix with strictly positive eigenvalues $\{\lambda_i\}_{i=1}^n$ and $\lambda_i<\lambda_1$ for all $i>1$.

### Part A

Show that $x^{(k)} = \frac{1}{\Vert A^k x^{(0)}\Vert} A^k x^{(0)}$ for all $k\geq 0$ and $A^k$ denotes the $k$-fold product of $A$ with itself.

### Part B

Let $A=U\Lambda U^T$ be a diagonalization of $A$ with $U$ an orthogonal matrix and $\Lambda$ a diagonal matrix. Show that $A^k = U\Lambda ^k U^T$.

### Part C

With $\{u_i\}_{i=1}^n$ an orthogonal basis of eigenvectors of $A$ from the columns of $U$, show that

$$
A^k x^{(0)} = \sum_{i=1}^n \lambda_i^k (u_i^Tx^{(0)})u_i
$$

and 

$$
\Vert A^k x^{(0)}\Vert = \lambda_1^k\sqrt{\vert u_1^T x^{(0)}\vert^2 + \sum_{i=2}^n \left(\frac{\lambda_i^k}{\lambda_1^k}\right)^2\vert u_i^T x^{(0)}\vert^2}.
$$

### Part D

Noting that $\frac{\lambda_i^k}{\lambda_1^k}\rightarrow 0$ monotonically as $k\rightarrow\infty$ if $i\geq 1$, show that $x^{(k)}\rightarrow u_1$ if $(u_1^Tx^{(0)})>0$ and $x^{(k)}\rightarrow -u_1$ if $(u_1^Tx^{(0)})<0$ as $k\rightarrow\infty$.

### Note: One can use this formulation to show that there is a constant $C$ and a $\delta\in (0,1)$ such that $\Vert x^{(k)}-u_1\Vert\leq C\delta^k$ when $x^{(k)}$ converges to $u_1$. When this happens, we will say that the convergence rate is *linear*.

## Solution to Problem 2

### Part A
Since $A$ has $n$ strictly positive eigenvalues, $A$ is full rank. That leads to $A^k$ in full rank. If $x^{(0)}$ is not zero vector, $A^kx^{(0)}$ would never be zero vector. So $x^{(k)}$ always exists.
### Part B
$$A^k=(U\Lambda U^T)(U\Lambda U^T)...(U\Lambda U^T)$$
$$=(U\Lambda^2U^T)...(U\Lambda U^T)$$
$$=...$$
$$=U\Lambda^kU^T$$
### Part C
Notice we have:
$$A^ku_i=A^{k-1}Au_i$$
$$=A^{k-1}\lambda_iu_i$$
$$=...$$
$$=\lambda_i^ku_i$$
Since $\{u_i\}_{i=1}^n$ a orthogonal basis of $R^n$, we have:
$$x^{(0)}=\sum_{i=1}^n(u_i^Tx^{(0)})u_i$$
Then:
$$A^kx^{(0)}=\sum_{i=1}^n(u_i^Tx^{(0)})A^ku_i$$
$$=\sum_{i=1}^n(u_i^Tx^{(0)})\lambda_i^ku_i$$
$$=\sum_{i=1}^n\lambda_i^k(u_i^Tx^{(0)})u_i$$
From the equation above, we have:
$$\Vert A^kx^{(0)}\Vert=\sqrt{\sum_{i=1}^{n}\lambda_i^{2k}\vert u_i^Tx^{(0)}\vert}$$
$$=\lambda_1^k\sqrt{\vert u_1^T x^{(0)}\vert^2 + \sum_{i=2}^n \left(\frac{\lambda_i^k}{\lambda_1^k}\right)^2\vert u_i^T x^{(0)}\vert^2}$$
### Part D
$$\frac{A^kx^{(0)}}{\Vert A^kx^{(0)}\Vert}=\frac{\sum_{i=1}^n \lambda_i^k (u_i^Tx^{(0)})u_i}{\lambda_1^k\sqrt{\vert u_1^T x^{(0)}\vert^2 + \sum_{i=2}^n \left(\frac{\lambda_i^k}{\lambda_1^k}\right)^2\vert u_i^T x^{(0)}\vert^2}}$$
$$=\frac{(u_1^Tx^{(0)})u_1+\sum_{i=2}^n (\frac{\lambda_i}{\lambda_1})^k (u_i^Tx^{(0)})u_i}{\sqrt{\vert u_1^T x^{(0)}\vert^2 + \sum_{i=2}^n \left(\frac{\lambda_i}{\lambda_1}\right)^{2k}\vert u_i^T x^{(0)}\vert^2}}$$
$$\rightarrow\frac{(u_1^Tx^{(0)})u_1}{\vert u_1^T x^{(0)}\vert},\ while\ k\rightarrow\infty$$
$$= \begin{cases}
u_1, &u_1^Tx^{(0)}>0 \\
-u_1, &u_1^Tx^{(0)}<0
\end{cases}
$$

## Problem 3

We have derived a **reduced singular value decomposition** in class. The full SVD of a matrix $A\in\mathbb{R}^{m\times n}$ is given by $A=U\Sigma V^T$ where $U\in\mathbb{R}^m$ and $V\in\mathbb{R}^n$ are orthogonal, and $\Sigma\in \mathbb{R}^{m\times n}$ has zero entries off its diagonal and has the singular values of $A$ on its principal diagonal.

If $A$ is an $n$ by $n$ matrix, then the full SVD of $A$ will look like $A= U\Sigma V^T$ where $U$, $\Sigma$, and $V$ are $n$ by $n$ matrices.

### Part A

An $n$ by $n$ matrix $A$ is said to be invertible if there is an $n$ by $n$ matrix $A^{-1}$ that satisfies

$$
A^{-1}A=AA^{-1}=I.
$$

If $rank(A)=n$, use the full singular value decomposition of $A$ to prove that $A$ is invertible.

### Part B

If $A$ is invertible, prove that $\det(A)\not=0$ using the multiplicative rule of determinants.

### Part C

If $\det(A)\not=0$, prove that $rank(A)=n$ using the full singular value decomposition of $A$ and the multiplicative rule of determinants.

### Note: This show that invertibility is equivalent to $rank(A)=n$, which is in turn equivalent to $\det(A)\not=0$.

## Solution to Problem 3
### Part A
Since $rank(A)=n$, $\Sigma$ has strictly positive diagonal elements. We write $\Sigma^{-1}$ as the diagonal matrix with all the diagonal elements being the reciprocal values of elements at the same positions of original matrix $\Sigma$.  
We can prove that $A^{-1}=V\Sigma^{-1}U^T$. That leads to:
$$A^{-1}A=V\Sigma^{-1}U^TU\Sigma V^T=I$$
$$AA^{-1}=U\Sigma V^TV\Sigma^{-1}U^T=I$$
### Part B
If $det(A)=0$, we have: $1=det(I)=det(AA^{-1})=det(A)det(A^{-1})=0$  
That leads to the contradiction.  
So $det(A)\neq 0$
### Part C
$$A=U\Sigma V^T$$
$$det(A)=det(U)det(\Sigma)det(V^T)$$
$$det(\Sigma)\neq0$$
That leads to all the diagonal elements of $\Sigma$ are strictly positive, which indicates $rank(A)=n$

```python
import pandas as pd
import numpy as np
from scipy import sparse
def load_movielen():
    File = pd.read_csv("u.data",names=['user_id','item_id','rating','timestamp'],sep='\t')
    df=pd.DataFrame(File)
    return sparse.csr_matrix((df.loc[:,'rating'],(df.loc[:,'user_id']-1,df.loc[:,'item_id']-1)),shape=(943,1682),dtype='float64').tolil()
ml_sparse_record=load_movielen()
```

```python
from scipy.sparse.linalg import svds
import numpy as np
rank=100
ml_sparse=ml_sparse.tocsc()
ml_sparse_raw=sparse.csc_matrix(ml_sparse)
U,s,Vt=svds(ml_sparse,k=rank)
S=np.zeros((100,100))
for i in range(rank):
    S[i,i]=s[i]
ml_sparse=np.dot(U,np.dot(S,Vt))

#for i in range(10):
#    templist=[]
#    for j in range(10):
#        templist.append(ml_sparse[i,j])
#    print(templist)
```
