## Problem 1

For $f\in C^1(\mathbb{R}^n)$ and $\gamma\in C^1(\mathbb{R};\mathbb{R}^n)$, the multivariate Taylor's theorem gives us a $\tau\in (t, t+h)$ such that 

$$
\gamma(t+h)=\gamma(t) + \frac{d\gamma}{dt}(\tau)h,
$$

and a $\nu$ between $\gamma(t)$ and $\gamma(t) + \frac{d\gamma}{dt}(\tau)h$ such that

$$
\begin{align}
f(\gamma(t+h))=&f\left(\gamma(t) + \frac{d\gamma}{dt}(\tau)h\right)\\
=&f(\gamma(t))+\nabla f(\nu)^T\frac{d\gamma}{dt}(\tau)h.
\end{align},
$$

Therefore,

$$
\frac{f(\gamma(t+h))-f(\gamma(t))}{h} =\frac{\nabla f(\nu)^T\frac{d\gamma}{dt}(\tau)h}{h}=\nabla f(\nu)^T\frac{d\gamma}{dt}(\tau)
$$

for any $h\not=0$. As $h\rightarrow 0$, $\tau\rightarrow t$, and $\nu\rightarrow \gamma(t)$. Consequently, the limit of this difference quotient is $\nabla f(\gamma(t))^T\frac{d\gamma}{dt}(t)$, and the first **multivariate chain rule** is

$$
\frac{d}{dt} f(\gamma(t)) = \nabla f(\gamma(t))^T\frac{d\gamma}{dt}(t)
$$

### Part A

If $g\in C^1(\mathbb{R}^m)$ and $f\in C^1(\mathbb{R}^n;\mathbb{R}^m)$, then $g\circ f:\mathbb{R}^n\rightarrow\mathbb{R}$ is in $C^1(\mathbb{R}^n)$. Prove the **chain rule**:

$$
\nabla(g\circ f)({\bf x}) = Df({\bf x})^T\nabla g(f({\bf x})).
$$


### Part B
If $g\in C^1(\mathbb{R}^k;\mathbb{R}^m)$ and $f\in C^1(\mathbb{R}^n;\mathbb{R}^k)$, then $g\circ f\in C^1(\mathbb{R}^n; \mathbb{R}^m)$. Prove the **chain rule**:

$$
D(g\circ f)({\bf x}) = Dg(f({\bf x}))Df({\bf x}).
$$


## Solution to Problem 1
### Part A
$$\nabla(g \circ f)({\bf x})=\begin{pmatrix}
\frac{\partial(g \circ f)}{\partial x_1}(x)\\
\vdots\\
\frac{\partial(g \circ f)}{\partial x_n}(x)
\end{pmatrix}$$
$$=\begin{pmatrix}
\nabla g(f(x))^T \frac{\partial f}{\partial x_1}(x)\\
\vdots\\
\nabla g(f(x))^T \frac{\partial f}{\partial x_n}(x)
\end{pmatrix}$$
$$= Df({\bf x})^T\nabla g(f({\bf x}))$$
### Part B
We write $g(y)$ as:
$$g(y)=\begin{pmatrix}
g_1(y)\\
\vdots\\
g_m(y)
\end{pmatrix}$$
in which $g_i(y)\in C^1(\mathbb{R}^k;\mathbb{R})$
Then:
$$D(g \circ f)(\bf x)=\begin{pmatrix}
\nabla(g_1 \circ f)(\bf x)^T)\\
\vdots\\
\nabla(g_m \circ f)(\bf x)^T)
\end{pmatrix}$$
$$=\begin{pmatrix}
\nabla g_1(f({\bf x}))^TDf(\bf x)\\
\vdots\\
\nabla g_m(f({\bf x}))^TDf(\bf x)
\end{pmatrix}$$
$$=\begin{pmatrix}
\nabla g_1(f({\bf x}))^T\\
\vdots\\
\nabla g_m(f({\bf x}))^T
\end{pmatrix}Df({\bf x})$$
$$=Dg(f({\bf x}))Df({\bf x})$$

## Problem 2

Suppose $y\in\mathbb{R}^N$, $X\in\mathbb{R}^{N\times(d+1)}$, and define $f:\mathbb{R}^{d+1}\rightarrow\mathbb{R}^N$ and $g:\mathbb{R}^N\rightarrow\mathbb{R}$ by 

$$
f(\beta)=X\beta-y\text{ and } g(x)=\frac{1}{2}\Vert x\Vert^2
$$

Compute $Df(\beta)$, $\nabla g(x)$, and use the chain rule to compute $\nabla (g\circ f)(\beta)=\nabla\left[\frac{1}{2}\Vert y-X\beta\Vert^2\right]$.


## Solution to Problem 2
We write $f(\beta)$ as:
$$\begin{pmatrix}
\alpha_1\\
\vdots\\
\alpha_N
\end{pmatrix}$$
We write elements of $X$ as $\{x_{i,j};i=1,...,N;j=1,...,d+1\}$
We write $\beta$ as:
$$\begin{pmatrix}
\beta_1\\
\vdots\\
\beta_{d+1}
\end{pmatrix}$$
We compute $X\beta$. Then we have:
$$X\beta=\begin{pmatrix}
\sum_{j=1}^{d+1}x_{1,j}\beta_j\\
\vdots\\
\sum_{j=1}^{d+1}x_{i,j}\beta_j\\
\vdots\\
\sum_{j=1}^{d+1}x_{N,j}\beta_j
\end{pmatrix}$$
Then we compute $D(f\beta)$:
$$D(f(\beta))=D(X\beta)$$
$$=\begin{pmatrix}
x_{1,1}&\cdots&x_{1,d+1}\\
x_{2,1}&\cdots&x_{2,d+1}\\
\vdots&\vdots&\vdots\\
x_{N,1}&\cdots&x_{N,d+1}
\end{pmatrix}$$
$$=X$$
Then we compute $\nabla g(x)$:
$$g(x)=\frac{1}{2}(x_1^2+x_2^2+\cdots+x_N^2)$$
$$\frac{\partial g(x)}{\partial x_i}=x_i,i=1,\cdots,N$$
That is:
$$\nabla g(x)=\begin{pmatrix}
x_1\\
\vdots\\
x_N
\end{pmatrix}$$
$$=x$$
$$\nabla(g\circ f)({\beta}) = Df({\beta})^T\nabla g(f({\beta}))$$
$$=X^T(X\beta-y)$$
$$=X^TX\beta-X^Ty$$

## Problem 3

Given $v\in\mathbb{R}^d$, let $f(x)=\frac{1}{2}\Vert x-v\Vert^2+\sum_{i=1}^d\vert x_i\vert$. Show that $f$ is strictly convex, and find the unique minimizer $x^\ast$ of $f$ over $\mathbb{R}^d$.

## Solution to Problem 3
We let $g(x)=\frac{1}{2}\Vert x-v\Vert^2$ and $h(x)=\sum_{i=1}^d\vert x_i\vert$
$$\frac{\partial^2 g(x)}{\partial x_i\partial x_j}=
\begin{cases}
1&when\ i=j\\
0&when\ i\neq j
\end{cases}$$
So $\nabla^2g(x)$ is an identity matrix, which is surely positive definite. That indicates $g(x)$ being strictly convex.  
As for $h(x)$, we suppose $x,y\in R^d$ and $t\in [0,1]$  
Then we have:
$$h(tx+(1-t)y)$$
$$=\sum_{i=1}^{d}\vert tx_i+(1-t)y_i\vert$$
$$\leq \sum_{i=1}^{d}(\vert tx_i\vert + \vert(1-t)y_i\vert)$$
$$=t\sum_{i=1}^{d}\vert x_i\vert+(1-t)\sum_{i=1}^{d}\vert y_i\vert$$
$$=th(x)+(1-t)h(y)$$
That indicates $h(x)$ being convex.  
So $f(x)$, the sum of $g(x)$ and $h(x)$, is strictly convex.  

We let $$s(q)=\min_p k(p,q)=\min_p(\frac{1}{2}(p-q)^2+\vert p\vert)$$
$$k(p,q)=\begin{cases}
\frac{1}{2}(p^2-(2q-2)p+q^2)&when\ p>0\\
\frac{1}{2}(p^2-(2q+2)p+q^2)&when\ p\leq0
\end{cases}$$
#### (1)
When $q<-1$:
$$min_{p,p\geq0}k(p,q)=\frac{1}{2}q^2$$
$$min_{p,p\leq0}k(p,q)=-q-\frac{1}{2}$$
We have:
$$\frac{1}{2}q^2\geq-q-\frac{1}{2}$$
So we have:
$$s(q)=min_{p}k(p,q)=-q-\frac{1}{2}\ (p=q+1)$$

#### (2)
When $-1\leq q\leq1$:
$$min_{p,p\geq0}k(p,q)=\frac{1}{2}q^2$$
$$min_{p,p\leq0}k(p,q)=\frac{1}{2}q^2$$
So we have:
$$s(q)=min_{p}k(p,q)=\frac{1}{2}q^2\ (p=0)$$

#### (3)
When $q>1$:
$$min_{p,p\geq0}k(p,q)=q-\frac{1}{2}$$
$$min_{p,p\leq0}k(p,q)=\frac{1}{2}q^2$$
We have:
$$q-\frac{1}{2}\leq \frac{1}{2}q^2$$
So we have:
$$s(q)=min_{p}k(p,q)=q-\frac{1}{2}\ (p=q-1)$$  
  
In conclusion:
$$s(q)=\begin{cases}
-q-\frac{1}{2}&when\ q<-1\\
\frac{1}{2}q^2&when\ -1\leq q\leq1\\
q-\frac{1}{2}&when\ q>1
\end{cases}$$
And we let $$r(q)=argmin_p k(p,q)=
\begin{cases}
q+1&when\ q<-1\\
0&when\ -1\leq q\leq1\\
q-1&when\ q>1
\end{cases}$$
  
Back to the problem, we have:
$$f(x)=\sum_{i=1}^{d}k(x_i,v_i)$$
$$\min_xf(x)=\min_x\sum_{i=1}^{d}k(x_i,v_i)$$
$$=\sum_{i=1}^{d}\min_{x_i}k(x_i,v_i),\text{ (considering the independence of }x_i)$$
$$=\sum_{i=1}^{d}s(v_i)$$
$$\text{with }x^\ast\text{ being }\begin{pmatrix}
r(v_1)\\
\vdots\\
r(v_d)
\end{pmatrix}$$

## Problem 4

$$
f(a,b,c) = -\log\det\begin{pmatrix} a & b\\b&c\end{pmatrix}.
$$

Show that $\nabla^2 f(a,b,c)$ is positive semidefinite if $a>0$ and $ac-b^2>0$. 


## Solution to Problem 4

We first compute $f(a,b,c)$:
$$f(a,b,c)=-log(ac-b^2)$$
We compute $\nabla f(a,b,c)$:
$$\nabla f(a,b,c)=\begin{pmatrix}
-\frac{c}{ac-b^2}\\
\frac{2b}{ac-b^2}\\
-\frac{a}{ac-b^2}
\end{pmatrix}$$
Then we compute $\nabla^2f(a,b,c)$:
$$\nabla^2f(a,b,c)=\begin{pmatrix}
\frac{c^2}{(ac-b^2)^2}&-\frac{2bc}{(ac-b^2)^2}&\frac{b^2}{(ac-b^2)^2}\\
-\frac{2bc}{(ac-b^2)^2}&\frac{2ac+2b^2}{(ac-b^2)^2}&-\frac{2ab}{(ac-b^2)^2}\\
\frac{b^2}{(ac-b^2)^2}&-\frac{2ab}{(ac-b^2)^2}&\frac{a^2}{(ac-b^2)^2}
\end{pmatrix}$$
$$=\frac{1}{(ac-b^2)^2}\begin{pmatrix}
c^2&-2bc&b^2\\
-2bc&2ac+2b^2&-2ab\\
b^2&-2ab&a^2
\end{pmatrix}$$
We have:
$$det(c^2)\geq 0$$
$$det\begin{pmatrix}
c^2&-2bc\\
-2bc&2ac+2b^2
\end{pmatrix}=2c^2(ac-b^2)\geq0$$
$$det\begin{pmatrix}
c^2&-2bc&b^2\\
-2bc&2ac+2b^2&-2ab\\
b^2&-2ab&a^2
\end{pmatrix}=(2a^2c^2+2b^4-4ab^2c)(ac-b^2)=2(ac-b^2)^2(ac-b^2)\geq0$$
So that indicates $\nabla^2f(a,b,c)$ is positive semidefinite.
